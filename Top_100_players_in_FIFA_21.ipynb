**Creating a list and replace the missing values**

```python
missing_values_list_FIFA_21 = ['pace','shooting','passing','dribbling','defending']
for x in missing_values_list_FIFA_21:
  FIFA_21[x].fillna(FIFA_21[x].min(), inplace=True)
```

```python
FIFA_21['physic'].fillna(FIFA_21['physic'].mean(), inplace=True)
```

```python
FIFA_21['player_traits'] = FIFA_21['player_traits'].fillna('No traits available')
```

```python
FIFA_21.drop(columns='defending_marking',inplace=True)
```

```python
missing_FIFA_21_data = FIFA_21.isnull().sum()
print(missing_FIFA_21_data[missing_FIFA_21_data > 0])
```

**1.AGE**

**2.COUNTRY**

```python
fig, ax1 = plt.subplots(figsize=(20,10))

graph = sns.countplot(ax=ax1, data=FIFA_21, x='nationality', order=FIFA_21['nationality'].value_counts().head(10).index, palette='Blues_r' )

ax1.set(xlabel='Count',ylabel='Country',title='Number of Top players per country',)
graph.set_xticklabels(graph.get_xticklabels(),rotation=90)

for p in graph.patches:
    height = p.get_height()
    graph.text(p.get_x()+p.get_width()/2.,height +0.1,height, ha="center")
    
plt.show()
```

**3. Player Position**

```python
FIFA_21['player_positions']
```

**With the modifications done in the list**

```python
positions_list=[]

[positions_list.append(x.split(',')[0]) for x in FIFA_21['player_positions'].values.tolist()]
    
FIFA_21['position']= positions_list


for x in FIFA_21['position']:
    if x in ['ST','CF','RW','LW']:
        FIFA_21['position'].replace(to_replace=x,value='Attacker',inplace=True) 
    if x in ['LM','RM','CM','CDM','CAM']:
        FIFA_21['position'].replace(to_replace=x,value='Midfielder',inplace=True)
    if x in ['LB','RB','CB']:
        FIFA_21['position'].replace(to_replace=x,value='Defender',inplace=True)
    if x in ['GK']:
        FIFA_21['position'].replace(to_replace=x,value='Goalie',inplace=True)
```

```python
FIFA_21['position']
```

**NO OF PLAYERS IN EACH POSITION**

```python
fig, ax1 = plt.subplots(figsize=(10,8))

graph = sns.countplot(ax=ax1, data=FIFA_21, x='position', order=FIFA_21['position'].value_counts().head(10).index, palette='Blues_r' )

ax1.set(xlabel='Count',ylabel='Position',title='Players count per position',)
graph.set_xticklabels(graph.get_xticklabels(),rotation=90)

for p in graph.patches:
    height = p.get_height()
    graph.text(p.get_x()+p.get_width()/2.,height +0.1,height, ha="center")
    
plt.show()
```
